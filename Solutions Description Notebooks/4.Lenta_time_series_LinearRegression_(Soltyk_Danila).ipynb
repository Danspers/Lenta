## Описание проекта

**Краткое описание:**
Необходимо создать алгоритм прогноза спроса на 14 дней для товаров собственного
производства. Гранулярность ТК-SKU-День.

**Термины:**
- ТК - Торговый комплекс
- SKU - Товар
- СП - собственное производство

**Задача:**
Построить модель прогноза спроса на основе мастер данных и данных продаж с учетом разных
признаков.

**Цель:**
Сгенерировать различные признаки и придумать интерпретируемую, описывающую правильные
зависимости (повышение цены вызывает логичное падение спроса), модель прогноза спроса.
Дальше необходимо сделать подневной прогноз спроса на тестовом периоде для каждого товара
и магазина, и команда Ленты оценит его качество в сравнении с свершившимся фактом.
Метрикой качества будет выступать WAPE, посчитанный на уровне товар, магазин, день. Если
есть пропущенные значения и по каким-то товарам не предоставлен прогноз, прогноз считается
равным нулю.

**Требования к результату:**
1. Файл в зафиксированном формате с результатом прогноза спроса(sales_submission.csv).
2. Воспроизводимый код на Python
3. Описание решения:
    + Описание модели прогноза спроса, которую вы зафитили
        + Признаки
        + интерпретация (`shapley values`),
        + кросс-валидация
        + алгоритмы
    + Описание вашего алгоритма оптимизации:
        + методология расчетов
        + скорость оптимизации

**Критерии оценки:**
У вас есть тестовый датасет за более поздние периоды в рамках групп товаров собственного
производства. Ваш алгоритм должен показать высокую точность прогнозирования спроса с
использованием признаков, которые вы сочтёте нужными.
Критерии оценки – WAPE + алгоритм прогноза спроса является понятным, эффективным и
масштабируемым.

```python
# функция метрики wape
def wape(y_true: np.array, y_pred: np.array):
    return np.sum(np.abs(y_true - y_pred)) / np.sum(np.abs(y_true))
```

2. Разделение всего датасета на два малых по признаку наличия промо-акции `pr_sales_type_id`.

```python
# контроль группировки
all = sales_df_train.shape[0]
dupls = sales_df_train[sales_df_train[['st_id', 'pr_sku_id', 'date']].duplicated(keep=False)].shape[0]
no_dupls = all - dupls

print('Исходное кол-во объектов:         ', all)
print('Кол-во неявных дубликатов:        ', dupls)
print('Кол-во объектов без дубликатов:   ', no_dupls)
print('Кол-во объектов после группировки:', no_dupls + dupls//2)
print('Группировка прошла успешно:       ', group_sales.shape[0] == (no_dupls + dupls//2))
```

```python
# создание новых столбцов
group_sales['promo_in_units'] = group_sales['pr_promo_sales_in_units'] / group_sales['pr_sales_in_units']
group_sales['promo_in_rub']   = group_sales['pr_promo_sales_in_rub']   / group_sales['pr_sales_in_rub']

group_sales['price'] = ((group_sales['pr_sales_in_rub'] * (1 - group_sales['promo_in_rub'])) /
                        (group_sales['pr_sales_in_units'] * (1 - group_sales['promo_in_units'])))

group_sales['promo_price'] = ((group_sales['pr_sales_in_rub'] * group_sales['promo_in_rub']) /
                              (group_sales['pr_sales_in_units'] * group_sales['promo_in_units']))

#group_sales.fillna(0, inplace=True)
group_sales.drop(['pr_promo_sales_in_units',
                  'pr_promo_sales_in_rub',
                  'promo_in_units',
                  'promo_in_rub'], axis=1, inplace=True)

display(group_sales)
```

Новые столбцы:
- `promo_in_units` - доля промо-товаров среди общего количества продаж.
- `promo_in_rub` - доля промо-выручки среди общего объёма денежных средств.
- `price` - цена товара в день продажи
- `promo_price` - цена товара по промо-акции

Это обязательное условие, поскольку высока вероятность, что алгоритмы будут воспринимать выбросы как закономерность, а не как исключение.

```python
# функция создание признаков (ver.2)
def make_time_features_v2(df, max_lag:int, window_size:int):
    
    # отстающие (предыдущие) признаки / lag features
    for lag in range (1, max_lag+1):
        df[f'pr_sales_in_rub_{lag}'] = df['pr_sales_in_units'].shift(periods=lag, fill_value=df.iloc[-1, lag-1])
    
    # скользящее среднее / rolling mean
    df['rolling_mean'] = df['pr_sales_in_units'].shift().rolling(window_size).mean()
    
    # удаление новообразованных пустых объектов в начале
    df.dropna(inplace=True)

    # календарные признаки / calendar features
    df['month']     = df.index.month
    df['dayofweek'] = df.index.dayofweek
    df['day']       = df.index.day
    #df['holiday']   = calendar['holiday']
    
```

Пояснения по выбранным факторам:

3) Фактор день: Нужен, поскольку этот фактор работает в паре с месяцем и дополняет его значимость.
4) Фактор день недели: ДА, т.к. присутствует недельная сезонность.
5) Факто lag: ДА, это параметр авторегрессии - учёт влияния вчерашних событий на сегодняшний день.
6) Фактор rolling: ДА, это аналог тренда. Чем больше окно сглаживания, тем ровнее линия тренда.
7) Фактор std: НЕТ, при реальном использовании в бизнесе нет возможности узнать разброс среднего в будущем.

### Прототип 4.0

Идея этого прототипа основывается на создание признаков из прошлого периода, продолжительность которого равна горизонту предсказывания. Если необходимо выполнить предсказание на следующие 2 недели от текущего дня, то создаётся целый перечень признаков из предыдущих 2 недель. В первую очередь это _отстающие_ (лаговые) и _скользящее_ признаки.

```python
# функция дополнение текущего периода данными из прошлого
def concat_with_past(sku_name:str, st_name:str):

    # тестовый датасет
    sku_present = sales_submission.loc[(sales_submission['pr_sku_id'] == sku_name) &
                                       (sales_submission['st_id'] == st_name),
                                       ['date', 'target']]
    sku_present.sort_values(by='date', inplace=True)
    sku_present.set_index('date', inplace=True)
    sku_present.columns = ['pr_sales_in_units']

    # исходный датасет
    sku_past = group_sales.loc[(group_sales['pr_sku_id'] == sku_name) &
                               (group_sales['st_id'] == st_name),
                               ['date', 'pr_sales_in_units']]
    sku_past.sort_values(by='date', inplace=True)
    sku_past.set_index('date', inplace=True)
    sku_past = sku_past.tail(len(sku_present)*2)

    # объединение и создание новых признаков
    return pd.concat([sku_past, sku_present])
```

```python
# функция создание признаков (ver.4)
def make_time_features_v4(df, max_lag:int, window_size:int, prediction_horizon=14-1):
    
    # отстающие (предыдущие) признаки / lag features
    for lag in range (1, max_lag+1):
        df[f'pr_sales_in_units_{lag}'] = df['pr_sales_in_units'].shift(periods = lag+prediction_horizon,
                                                                       fill_value = df.iloc[-1, lag-1])
    
    # скользящее признаки / rolling features
    df['rolling_mean'] = df['pr_sales_in_units_1'].shift().rolling(window_size).mean()
    df['rolling_std']  = df['pr_sales_in_units_1'].shift().rolling(window_size).std()
    
    # удаление новообразованных пустых объектов в начале
    df.dropna(inplace=True)

    # календарные признаки / calendar features
    df['month']     = df.index.month
    df['dayofweek'] = df.index.dayofweek
    df['day']       = df.index.day
    #df['holiday']   = calendar['holiday']
    
```

```python
# циклическая обработка
models = []
for i in tqdm(range(len(monotonic_time_series))):

    # выборка данных для одного единственного объекта (временного ряда)
    st_name  = monotonic_time_series.loc[i, 'st_id']
    sku_name = monotonic_time_series.loc[i, 'pr_sku_id']
    sku = group_sales.loc[(group_sales['st_id'] == st_name) &
                          (group_sales['pr_sku_id'] == sku_name),
                          ['date', 'pr_sales_in_units']]

    # дополнение/удаление признаков
    sku.set_index('date', inplace=True)
    make_time_features_v4(sku, max_lag=4, window_size=3)

    # выделение признаков
    target = sku['pr_sales_in_units']
    features = sku.drop(['pr_sales_in_units'], axis=1)

    # выделение валидационной выборки
    X_train, X_valid, y_train, y_valid = train_test_split(features,
                                                          target,
                                                          shuffle = False,
                                                          test_size = 0.1,
                                                          random_state = RANDOM_STATE)

    # масштабирование признаков
    numeric = ['pr_sales_in_units_1', 'pr_sales_in_units_2',
               'pr_sales_in_units_3', 'pr_sales_in_units_4',
               'rolling_mean', 'rolling_std']     
    scaler = StandardScaler()
    scaler.fit(X_train[numeric])
    for features in [X_train, X_valid]:
        features[numeric] = scaler.transform(features[numeric])
    
    # Линейная регрессия
    LR_model = LinearRegression()
    LR_model.fit(X_train, y_train)
    prediction = LR_model.predict(X_valid)
    wape_valid = wape(y_valid, prediction)

    # Dummy-модель
    dummy_median = DummyRegressor(strategy='median')
    dummy_median.fit(X_train, y_train)
    dummy_prediction = dummy_median.predict(X_valid)
    dummy_wape_valid = wape(y_valid, dummy_prediction)

    # дополнение таблицы значением метрики предсказания
    if wape_valid <= dummy_wape_valid:
        monotonic_time_series.loc[i, 'wape'] = wape_valid
        monotonic_time_series.loc[i, 'model'] = LR_model
    else:
        monotonic_time_series.loc[i, 'wape'] = dummy_wape_valid
        monotonic_time_series.loc[i, 'model'] = dummy_median

# качество предсказаний индивидуальных вр. рядов
monotonic_time_series['wape, %'] = monotonic_time_series['wape'] * 100
display(monotonic_time_series['wape, %'].to_frame().head(10))
```
